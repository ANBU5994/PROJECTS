```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import os
import csv
```

```python
# loading the diabetes dataset to a pandas DataFrame
heart_dataset = pd.read_csv("D:\\MP 1\\MODEL 2\\heart.csv")
```

```python
# separating the data and labels
X = heart_dataset.drop(columns = 'target', axis=1)
Y = heart_dataset['target']
```

Data Standardization

```python
scaler = StandardScaler()
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python

# Get patient name and ID
patient_name = input("Enter Patient Name: ")
patient_id = input("Enter Patient ID: ")

# Initialize an empty tuple for user input
input_data = ()
statements = ["age", "Gender (Male = 1, Female = 0)", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "sloc", "ca", "thal"]

# Get user input for each element
for i in range(0, 13):
    user_input = float(input(f"Enter the {statements[i]}: "))
    input_data += (user_input,)

# Display the user-entered data
print("User entered data:", input_data)

# Convert input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array for a single instance prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the input data (Assuming scaler is defined and fitted)
std_data = scaler.transform(input_data_reshaped)
print("Standardized data:", std_data)

# Make a classification prediction (Assuming classifier is defined and trained)
heart_status_pred = classifier.predict(std_data)
5
# Define the outcome based on prediction (Example: 1 for Diabetic, 0 for Non-Diabetic)
if heart_status_pred == 1:
    heart_status = 'The person Have Heart disease'
    health_scheme = 'Ayushman Bharat Pradhan Mantri Jan Arogya Yojana (AB-PMJAY)'
else:
    heart_status = 'The person Non Heart Disease'
    health_scheme = 'Scheme not to be needed'

# Print the diabetic status and health scheme recommendation
print(heart_status)
print(health_scheme)

# Create a DataFrame with the data
data_dict = {
    "Patient Name": [patient_name],
    "Patient ID": [patient_id],
    "age": [input_data[0]],
    "Gender": [input_data[1]],
    "cp": [input_data[2]],
    "trestbps": [input_data[3]],
    "chol": [input_data[4]],
    "fbs": [input_data[5]],
    "restecg": [input_data[6]],
    "thalach": [input_data[7]],
    "exang": [input_data[8]],
    "oldpeak": [input_data[9]],
    "sloc": [input_data[10]],
    "ca": [input_data[11]],
    "thal":[input_data[12]],
    "heart Status": [heart_status],
    "Health Scheme": [health_scheme]
}

df = pd.DataFrame(data_dict)

# Define the file path to store the data
file_path ="D:\\MP 1\\PATIENT DETAILS - HEART.xlsx"

# Check if the file exists, if so, append the data
if os.path.exists(file_path):
    # Load the existing data
    existing_df = pd.read_excel(file_path)
    # Append the new data
    df = pd.concat([existing_df, df], ignore_index=True)

# Save to Excel
df.to_excel(file_path, index=False)

print(f"Data successfully saved to {file_path}")
```
